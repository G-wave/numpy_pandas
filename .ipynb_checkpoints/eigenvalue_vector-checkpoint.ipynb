```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.mat([[3,1],[1,3]])  # lambda_1=4 x1=[1,1]; lambda_2 = 2 x2 = [-1,1]
```

```python
x1 = np.mat([[1],[1]])
x2 = np.mat([[-1],[1]])
```

```python
plt.plot((0,x1[0]),(0,x1[1]))
plt.plot((0,x2[0]),(0,x2[1]))
plt.show()
```

```python
plt.plot((0,(A*x1)[0][0]),(0,(A*x1)[1][0]))
plt.plot((0,(A*x2)[0][0]),(0,(A*x2)[1][0]))
plt.show()
```
